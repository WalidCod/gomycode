```python
#Correlation map between all features 
def plot_correlation_map( dataset1 ):

    corr = dataset1.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(dataset1)
```

```python
#plot correlation map
#the map shows correlation coefficients between all features, all values are included between -1 and 1, by analysing the map we can get a clear idea about the relationships between different features and between features and the target.  
#The correlation map helps us to get a clear idea about relationships between features, in fact The closer corr value is to +1 or -1, the more closely the two variables are related. if correlation cpefficient is close to zero we deduct that features are not related 
#if coefficient value is positive that means if one feature value gets larger the other gets larger too, if it is negative then when one feature value gets larger the other one gets smaller
```
